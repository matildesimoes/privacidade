```python
import matplotlib.pyplot as plt

# Data
naive_hashing_sizes = [100, 10000, 50000, 100000]
naive_hashing_times = [0.0, 0.1, 0.6, 1.0]

dh_based_sizes = [100, 10000, 50000, 100000]
dh_based_times = [0.1, 10.1, 61.6, 126.8]

ot_based_sizes = [100, 10000, 50000, 100000]
ot_based_times = [0.4, 0.4, 0.7, 0.9]

# Plot
plt.figure(figsize=(10, 6))

plt.plot(naive_hashing_sizes, naive_hashing_times, marker='o', label='Naive Hashing')
plt.plot(dh_based_sizes, dh_based_times, marker='s', label='Baseado em DH')
plt.plot(ot_based_sizes, ot_based_times, marker='^', label='Baseado em OT')

# Labels and title
plt.xlabel('N de Elementos')
plt.ylabel('Tempo Necessrio (s)')
plt.title('Tempo Necessrio vs N de Elementos para Cada Protocolo')
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
naive_hashing_sizes = [100, 10000, 50000, 100000]
naive_hashing_data = [0.0, 0.2, 0.8, 2.0]

dh_based_sizes = [100, 10000, 50000, 100000]
dh_based_data = [0.0, 0.11, 5.1, 10.1]

ot_based_sizes = [100, 10000, 50000, 100000]
ot_based_data = [0.0, 0.11, 5.0, 10.2]

# Plot
plt.figure(figsize=(10, 6))

plt.plot(naive_hashing_sizes, naive_hashing_data, marker='o', label='Naive Hashing')
plt.plot(dh_based_sizes, dh_based_data, marker='s', label='Baseado em DH')
plt.plot(ot_based_sizes, ot_based_data, marker='^', label='Baseado em OT')

# Labels and title
plt.xlabel('N de Elementos')
plt.ylabel('Total de Dados Enviados e Recebidos (s)')
plt.title('Total de Dados vs N de Elementos para Cada Protocolo')
plt.legend()
plt.grid(True)

# Show plot
plt.tight_layout()
plt.show()
```
